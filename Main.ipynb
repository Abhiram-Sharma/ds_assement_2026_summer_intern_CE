```python
import pandas as pd
import numpy as np
from scipy import stats

def detect_outliers(df, method='IQR', threshold=1.5, z_thresh=3.0, verbose=True):
    """
    Perform outlier detection on numeric columns of a DataFrame.
    Automatically converts numeric-looking object columns.
    
    Parameters:
        df (pd.DataFrame): Input DataFrame.
        method (str): 'IQR' or 'Z-score'.
        threshold (float): IQR multiplier (default 1.5).
        z_thresh (float): Z-score threshold (default 3.0).
        verbose (bool): Print debug info.
    
    Returns:
        pd.DataFrame: Boolean DataFrame marking outliers.
    """
    # Try converting object columns that look numeric
    df_cleaned = df.copy()
    for col in df_cleaned.select_dtypes(include='object').columns:
        try:
            df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='ignore')
        except Exception:
            pass  # leave non-convertible columns as-is

    df_numeric = df_cleaned.select_dtypes(include=[np.number])
    if verbose:
        print("Numeric columns detected:", list(df_numeric.columns))
    
    if df_numeric.empty:
        if verbose:
            print("No numeric columns found. Skipping outlier detection.")
        return pd.DataFrame(index=df.index)
    
    outliers = pd.DataFrame(index=df.index)

    if method == 'IQR':
        for col in df_numeric.columns:
            Q1 = df_numeric[col].quantile(0.25)
            Q3 = df_numeric[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - threshold * IQR
            upper_bound = Q3 + threshold * IQR
            outliers[col + '_outlier'] = ~df_numeric[col].between(lower_bound, upper_bound)
    elif method == 'Z-score':
        z_scores = np.abs(stats.zscore(df_numeric, nan_policy='omit'))
        if isinstance(z_scores, np.ndarray):
            z_scores = pd.DataFrame(z_scores, columns=df_numeric.columns, index=df.index)
        for col in df_numeric.columns:
            outliers[col + '_outlier'] = z_scores[col] > z_thresh
    else:
        raise ValueError("Method must be 'IQR' or 'Z-score'")
    
    return outliers
```

```python
import math

def preprocess_battery_data(battery_df):
    """
    Converts the battery data from a list of records into a dictionary
    for efficient lookups.

    Args:
        battery_df: A list of dictionaries, where each dictionary represents
                    a battery reading record. Each record must have 'VID',
                    'TS_parsed', and 'Battery_Level'.

    Returns:
        A dictionary where each key is a 'VID' and each value is a sorted
        list of (timestamp, battery_level) tuples.
        Example: {'vid_1': [(ts1, 99.8), (ts2, 99.7), ...]}
    """
    battery_lookup = {}
    for row in battery_df:
        vid = row['VID']
        ts = row['TS_parsed']
        level = row['BATTERY_LEVEL']
        
        # If the VID is not yet in our dictionary, initialize it with an empty list
        if vid not in battery_lookup:
            battery_lookup[vid] = []
            
        # Append the (timestamp, level) tuple. The data is pre-sorted,
        # so this list will remain sorted by timestamp.
        battery_lookup[vid].append((ts, level))
        
    return battery_lookup

def find_closest_battery_level(vid, event_ts, battery_lookup, time_threshold_seconds):
    """
    Finds the closest battery level for a given event, using binary search.

    Args:
        vid: The vehicle ID for the event.
        event_ts: The timestamp of the event.
        battery_lookup: The pre-processed dictionary of battery data.
        time_threshold_seconds: The maximum allowed time difference (in seconds)
                                for a valid match.

    Returns:
        The battery level if a valid match is found within the threshold,
        otherwise None.
    """
    # Step 1: Check if we have any battery data for this vehicle ID.
    if vid not in battery_lookup:
        return None
        
    vehicle_battery_readings = battery_lookup[vid]
    
    # We only need the timestamps for the binary search.
    battery_timestamps = [reading[0] for reading in vehicle_battery_readings]

    # Step 2: Perform a binary search to find the "insertion point" for the
    # event timestamp in the sorted list of battery timestamps. This is the
    # most efficient way to find the nearest neighbors.
    
    low, high = 0, len(battery_timestamps) - 1
    # This will hold the index of the best candidate found so far.
    best_match_idx = -1

    # Standard binary search loop
    while low <= high:
        mid = (low + high) // 2
        if battery_timestamps[mid] == event_ts:
            best_match_idx = mid
            break # Exact match found, no need to search further.
        elif battery_timestamps[mid] < event_ts:
            low = mid + 1
        else:
            high = mid - 1

    # Step 3: If no exact match was found, the closest timestamps will be
    # neighbors of the final 'low' index from the binary search.
    if best_match_idx == -1:
        # 'low' is the insertion point. The candidates are at 'low' and 'low - 1'.
        # We must check if these indices are valid.
        idx1 = low - 1
        idx2 = low

        # Determine which of the two candidates is closer.
        diff1 = abs(event_ts - battery_timestamps[idx1]) if idx1 >= 0 else float('inf')
        diff2 = abs(event_ts - battery_timestamps[idx2]) if idx2 < len(battery_timestamps) else float('inf')

        if diff1 < diff2:
            best_match_idx = idx1
        else:
            best_match_idx = idx2

    # Step 4: Validate the best match against the time threshold.
    if best_match_idx != -1 and best_match_idx < len(battery_timestamps):
        closest_ts = battery_timestamps[best_match_idx]
        time_difference = abs(event_ts - closest_ts)
        time_difference_in_seconds = time_difference.total_seconds()
        
        if time_difference_in_seconds <= time_threshold_seconds:
            # If valid, return the corresponding battery level.
            return vehicle_battery_readings[best_match_idx][1]

    # If no suitable match was found within the threshold.
    return None

def merge_events_with_battery_data(events_df, battery_df):
    """
    Main function to orchestrate the merging process.

    Args:
        events_df: The list of event records.
        battery_df: The list of battery records.

    Returns:
        A new list of 'EVENTS_Final' records, with the 'Battery_Level' column added.
    """
    print("Step 1: Pre-processing battery data for fast lookups...")
    battery_lookup = preprocess_battery_data(battery_df)
    print("Done.\n")
    
    EVENTS_Final = []
    time_threshold = 300 # +/- 300 seconds

    print("Step 2: Iterating through events and finding matching battery levels...")
    for event in events_df:
        vid = event['VID']
        event_ts = event['TS_parsed']
        
        # Find the battery level using our efficient search function.
        battery_level = find_closest_battery_level(vid, event_ts, battery_lookup, time_threshold)
        
        # Create the new record for the final dataframe.
        final_event_record = {
            'VID': vid,
            'TS': event['TS'],
            'TYPE': event['TYPE'],
            'VAL': event['VAL'],
            'BATTERY_LEVEL': battery_level # This will be None if no match was found
        }
        EVENTS_Final.append(final_event_record)
    print("Done.\n")
    
    return EVENTS_Final
```
